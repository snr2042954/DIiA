```python
import pandas as pd

# Define the file path (two folders back, then into 'Data' folder)
file_path = "../../Data/clg17zbrv0006l50f59exvdsm.csv"

# Try to read the CSV file
try:
    # Read the CSV file
    data = pd.read_csv(file_path)
    # Display the first few rows of the dataset
    print("data loaded")
except FileNotFoundError:
    print(f"File not found at path: {file_path}. Please check the file name and path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd
import json
```

```python
# Initialize an empty list to collect the rows from vangoghbrabant.com
filtered_rows = []

for index, row in data.iterrows():
    meta_value = row['meta']
    if isinstance(meta_value, str):  # Ensure it's a string
        try:
            # Parse the 'meta' string as JSON
            meta_dict = json.loads(meta_value)
            # Check if the 'source' key contains "vangoghbrabant.com"
            if 'source' in meta_dict and 'vangoghbrabant.com' in meta_dict['source']:
                # Append the entire row to the list
                filtered_rows.append(row)
        except json.JSONDecodeError:
            pass  # Skip invalid JSON strings

# Create a new DataFrame from the filtered rows
vangoghbrabant_df = pd.DataFrame(filtered_rows)
```

```python
# Display the filtered DataFrame
vangoghbrabant_df
```

```python
# Export the filtered DataFrame to a CSV file
output_file_path = "../../Data/vangoghbrabant_filtered.csv"
vangoghbrabant_df.to_csv(output_file_path, index=False)
```
